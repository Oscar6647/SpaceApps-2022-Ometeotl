```python
df = df.drop("LLuvia", axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix, classification_report, precision_score
```

```python
X = df.drop("PredLluvia",axis=1)
y = df["PredLluvia"]
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
dt = DecisionTreeClassifier()  
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
```

```python
accuracy = accuracy_score(
            y_true    = y_test, 
            y_pred    = y_pred, 
            normalize = True)
print(f"Modelo DecisionTreeClassifier")
print("")
print(f"Test size: 0.2")
print("")
print(f"El accuracy test del modelo DecisionTreeClassifier: {100*accuracy}%")

precision = precision_score(
            y_true    = y_test,
            y_pred    = y_pred,
            average = "micro"
           )
print("")
print(f"El test de precisn es de: {100*precision}%")
print("")
cf = confusion_matrix(
            y_true    = y_test,
            y_pred    = y_pred,
            normalize = "true")
sensitivityTP = cf[0,0]/(cf[0,0]+cf[0,1])
print('Sensitividad',sensitivityTP)
specificity = cf[1,1]/(cf[1,0]+cf[1,1])
print("")
print('Especificidad',specificity)
```

```python
dt = RandomForestClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
```

```python
accuracy = accuracy_score(
            y_true    = y_test, 
            y_pred    = y_pred, 
            normalize = True)
print(f"Modelo RandomForestClassifier")
print("")
print(f"Test size: 0.2")
print("")
print(f"El accuracy test del modelo RandomForestClassifier: {100*accuracy}%")

precision = precision_score(
            y_true    = y_test,
            y_pred    = y_pred,
            average = "micro"
           )
print("")
print(f"El test de precisn es de: {100*precision}%")
print("")
cf = confusion_matrix(
            y_true    = y_test,
            y_pred    = y_pred,
            normalize = "true")
sensitivityTP = cf[0,0]/(cf[0,0]+cf[0,1])
print('Sensitividad',sensitivityTP)
specificity = cf[1,1]/(cf[1,0]+cf[1,1])
print("")
print('Especificidad',specificity)
```

## CONTAMINANTES EN MONTERREY

```python
cd "/content/gdrive/MyDrive/SpaceApps/Ambiente"
```

```python
data = pd.read_csv("ContAMM.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data = data.drop(['LOCALIDAD',"AO"], axis=1)
```

```python
data.head()
```

```python
data = data.assign(PredDanger=0)
data.head()
```

```python
data['PM2.5'].fillna(27, inplace = True)
data.head()
```

```python
data["SO2"]= data["SO2"].replace(['D.I.','N.D.'],[0.007,0.006])
data['SO2'].fillna(0.009, inplace = True)
data.tail()
```

```python
data["CO"]= data["CO"].replace(['N.D.'],[1.37])
data['CO'].fillna(random.random()+ 1, inplace = True)
data.tail()
```

```python
data['O3'].fillna(0.090, inplace = True)
data.tail()
```

```python
data["NO2"]= data["NO2"].replace(['D.I.','N.D.'],[0.030,0.035])
data['NO2'].fillna(0.033, inplace = True)
data.tail()
```

Se buscr predecir si en los siguientes aos, los niveles de contaminantes llegan a niveles que puedan ser dainos para las personas en el Area Metropolitana de Monterrey

Se usar la columna "PredDanger" como variable predictora dependiente, y sus valores se basarn en si los niveles de los otros contaminantes llegan a un nivel peligroso para poder salir afuera. 

En este caso, se definir como "peligroso" cuando el nivel de PM10 sea mayor o igual a 100 o cuando el intervalo de O3, NO2, SO2 y CO sean mayores o iguales a 0.071, 0.211, 0.111 y 11 respectivamente (Esto, con base en los datos proporcionados por AIRE NL) 

```python
data.loc[data.PM10>=100,'PredDanger']=1
data.head(20)
```

```python
data["CO"].unique()
```

```python
data['CO'] = pd.to_numeric(data['CO'])
data['O3'] = pd.to_numeric(data['O3'])
data['NO2'] = pd.to_numeric(data['NO2'])
data['SO2'] = pd.to_numeric(data['SO2'])
```

```python
data.loc[data.O3>=0.71,'PredDanger']=1
```

```python
data.loc[data.NO2>=0.211,'PredDanger']=1
```

```python
data.loc[data.CO>=11,'PredDanger']=1
```

```python
data.loc[data.SO2>=0.111,'PredDanger']=1
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix, classification_report, precision_score
```

```python
X = data.drop("PredDanger",axis=1)
y = data["PredDanger"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
dt = RandomForestClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
```

```python
accuracy = accuracy_score(
            y_true    = y_test, 
            y_pred    = y_pred, 
            normalize = True)
print(f"Modelo RandomForestClassifier")
print("")
print(f"Test size: 0.2")
print("")
print(f"El accuracy test del modelo RandomForestClassifier: {100*accuracy}%")

precision = precision_score(
            y_true    = y_test,
            y_pred    = y_pred,
            average = "micro"
           )
print("")
print(f"El test de precisn es de: {100*precision}%")
print("")
cf = confusion_matrix(
            y_true    = y_test,
            y_pred    = y_pred,
            normalize = "true")
sensitivityTP = cf[0,0]/(cf[0,0]+cf[0,1])
print('Sensitividad',sensitivityTP)
specificity = cf[1,1]/(cf[1,0]+cf[1,1])
print("")
print('Especificidad',specificity)
```
